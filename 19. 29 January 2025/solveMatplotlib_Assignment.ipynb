```python
# We need data to work with, lets create data first!
import numpy as np
x = np.arange(0,10)
y = x*2
z = x**2
z
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
#plt.show() for non-notebook users
```

# Question: 1. Instructions:
* ** Use <code>plt.figure()</code> and create a figure object, say fig.**
* ** Add an axis to the figure canvas (`fig`), `[left, bottom, width, height] is [0,0,1,1]`. <i>Hint: <code>add_axes()</code><i>**
* ** Plot the data x,y on the axes**
* ** Set the labels x and y**
* ** Set the title**
* ** Create legends as shown in the plot**

Match the output below: 

```python
# Exercise 1 Code
fig = plt.figure()
ax_minhaz = fig.add_axes([0, 0, 1, 1])  # [left, bottom, width, height]
ax_minhaz.plot(x, y, label="Color")
ax_minhaz.set_xlabel('x')
ax_minhaz.set_ylabel('y')
ax_minhaz.set_title('Line Plot of x vs y')
ax_minhaz.legend()
plt.show()
```

# Question: 2. Instructions
* ** Create a figure object**
* ** Put two axes on the figure at [0,0,1,1] and [0.1,0.6,0.2,0.2]**
* ** Plot data (x,z) on main axes**
* ** Plot data (x,z) on the small axes and set its x limit b/w 1, 2, y limit 0, 5**
* ** Replicate the figure to axes, titles, legends etc**

```python
fig = plt.figure()

# Main axes
ax_main = fig.add_axes([0, 0, 1, 1])
ax_main.plot(x, z, "b*-", label='z = x')
ax_main.set_xlabel('x')
ax_main.set_ylabel('z')
ax_main.set_title('Main Plot')
ax_main.legend()

# Inset axes
ax_inset = fig.add_axes([0.1, 0.6, 0.2, 0.2])  # [left, bottom, width, height]
ax_inset.plot(x, z,"ro-", label='Inset')
ax_inset.set_xlim(1, 2)
ax_inset.set_ylim(0, 5)
ax_inset.set_title('Zoomed Inset')
ax_inset.legend()

plt.show()
```

# Question: 3. Replicate the output using <code>subplots()</code> and for loop** -- Note, the data is x and y. 

```python
# Exercise 3 Code (using subplots() and for loop)
fig, axes = plt.subplots(1, 3)  # 1 row, 3 columns
for ax in axes:
    ax.plot(x, y, 'b', label="col Name")
    ax.set_xlabel('X_label')
    ax.set_ylabel('Y_label')
    ax.set_title('Title')
    ax.legend()
    plt.tight_layout()
plt.show()
```

# Question:4. Replicate the output using <code>subplots()</code>

```python
# Exercise 4 Code (using subplots())
fig, axes = plt.subplots(1, 3)  # 1 row, 3 columns

# Plot on each subplot
axes[0].plot(x, y, "r*-", label = "subplot01")
axes[0].set_title('y = 2x')
axes[0].legend()

axes[1].plot(x, y, "g*-", label = "subplot02")
axes[1].set_title('y = 2x')
axes[1].legend()

axes[2].plot(x, y, "b*-", label = "subplot03")
axes[2].set_title('y = 2x')
axes[2].legend()

plt.tight_layout()
plt.show()
```

# Question:5. plot `(x,y)`, `(x,z)` and `(x,y)` on the above empty canvas to replicate the plot given below.**

```python
# Exercise 5 Code
plt.figure()
plt.plot(x, y, "rp-", label = "subplot01")
plt.plot(x, z, "g--", label = "subplot02")
plt.plot(x, y, "bx-", label = "subplot03")
plt.xlabel('x')
plt.ylabel('y/z')
plt.title('Multiple Line Plots')
plt.legend()
plt.show()
```

# Question:6. Resize the above plots to `figsize = (3,4)`

```python
# Exercise 6 Code (resize to figsize=(3,4))
plt.figure(figsize=(3, 4))
plt.plot(x, y, "rp-", label = "subplot01")
plt.plot(x, z, "g--", label = "subplot02")
plt.plot(x, y, "bx-", label = "subplot03")
plt.xlabel('x')
plt.ylabel('y/z')
plt.title('Multiple Line Plots')
plt.legend()
plt.show()
```
